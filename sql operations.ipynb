```python
#3.3
result=cursor.execute("SELECT STATE_UT, DISTRICT, year,Rape FROM crimes_stats GROUP BY STATE_UT, DISTRICT, year,Rape HAVING MIN(Rape) ORDER BY Rape")
print("Minimum Rape = ", result.fetchone())
```

```python
#3.2
result=cursor.execute("SELECT STATE_UT, DISTRICT, year,Rape FROM crimes_stats GROUP BY STATE_UT, DISTRICT, year,Rape HAVING MAX(Rape) ORDER BY MAX(Rape) DESC")
print("Maximum Rape = ", result.fetchone())
```

```python
#3.2
result=cursor.execute("SELECT STATE_UT, DISTRICT, year,Kidnapping_Abduction FROM crimes_stats GROUP BY STATE_UT, DISTRICT, year,Kidnapping_Abduction  HAVING MAX(Kidnapping_Abduction) ORDER BY MAX(Kidnapping_Abduction) DESC")
print("Maximum Kidnapping_Abduction = ", result.fetchone())
```

```python
#3.3
result=cursor.execute("SELECT STATE_UT, DISTRICT, year,Kidnapping_Abduction FROM crimes_stats GROUP BY STATE_UT, DISTRICT, year,Kidnapping_Abduction  HAVING MIN(Kidnapping_Abduction ) ORDER BY Kidnapping_Abduction ")
print("Minimum Kidnapping_Abduction  = ", result.fetchone())
```
